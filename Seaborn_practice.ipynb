```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('iris.csv')
```

```python
data.head(5)
```

```python
plt.scatter('sepal.length','variety',data=data)
plt.show()
```

```python
x=data['sepal.length']
y=data['sepal.width']
plt.scatter(x,y)
sns.set()
sns.set_style('whitegrid')
plt.show()
```

```python
x=data['petal.length']
y=data['petal.width']
plt.scatter(x,y)
sns.set_style('darkgrid')
plt.show()
```

```python
x=data['variety']
y=data['petal.width']
plt.scatter(x,y)
sns.set_style('whitegrid')
sns.despine()
plt.show()
```

```python
x=data['petal.length']
y=data['petal.width']
plt.scatter(x,y)
sns.set_style('whitegrid', {'grid.colour':'.8'})
plt.show()
```

```python
x=data['variety']
y=data['petal.width']
plt.scatter(x,y)
sns.set_style('whitegrid',{'grid.color':'.8','grid.linestyle':":"})
plt.show()
```

```python
x=data['petal.length']
y=data['variety']
plt.scatter(x,y)
sns.set_style('darkgrid')
sns.set_context('notebook')
plt.show()
```

```python
x=data['sepal.width']
y=data['variety']
plt.scatter(x,y)
sns.set_style('whitegrid',{'grid.color':".2","grid.linestyle":":"})
sns.set_context('poster')
plt.show()
```

```python
# Relational Plot
sns.relplot(data=data,x='variety',y='sepal.length')
sns.set_style('whitegrid',{"grid.color":".01","grid.linestyle":":"})
sns.set_context('poster')
```

```python
sns.relplot(data=data,x='sepal.length',y='sepal.width')
sns.set_context('notebook')
```

```python
sns.relplot(data=data,x='petal.length',y="petal.width",hue='variety')
```

```python
sns.relplot(data=data,x='sepal.length',y='sepal.width',hue='petal.width',col='variety')
```

```python
sns.relplot(data=data,x='petal.length',y='sepal.length',hue='sepal.width',col='variety',col_wrap=3)
```

```python
sns.relplot(data=data,x='petal.length',y="petal.width",kind="line")
```

```python
sns.relplot(data=data,x="petal.length",y="petal.width",kind="line",errorbar=None)
```

```python
# Point Plot
sns.pointplot(data=data,x='sepal.length',y='sepal.width')
```

```python
x=data['petal.length'].iloc[:10]
y=data['petal.width'].iloc[:10]
sns.pointplot(data=data,x=x,y=y)
plt.show()
```

```python
x=data['petal.length'].iloc[:150]
y=data['petal.width'].iloc[:150]
sns.pointplot(data=data,x=x,y=y,hue='variety',palette='Accent')
```

```python
# Joint Plot
sns.set_style('darkgrid')
sns.jointplot(data=data,x='petal.length',y="petal.width")
```

```python
sns.jointplot(data=data,x='sepal.length',y='sepal.width',kind='reg')
```

```python
sns.jointplot(data=data,x='sepal.length',y='sepal.width',kind="kde")
```

```python
sns.jointplot(data=data,x='sepal.length',y='sepal.width',kind="hex")
```

```python
# Reg Plot
sns.regplot(data=data,x='sepal.length',y='sepal.width')
```

```python
# Box Plot
sns.boxplot(data=data,x='petal.length',y='petal.width')
```

```python
x=data['sepal.length'].iloc[:10]
y=data['sepal.width'].iloc[:10]
sns.boxplot(data=data,x=x,y=y)
```

```python
# Correlational Matrix
# if 1 => positive relation, if -1 => negative correlation ,  if 0 => no correlation
corr=data.corr()
sns.heatmap(corr,annot=True,linewidths=.5,cmap="YlGn")
```

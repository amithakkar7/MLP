```python
categories=np.array(['A','B','C','D']) 
values=np.array([3,8,1,10])
values2=np.array([5,6,2,9])
values3=np.array([15,36,48,29])
bar_width=0.35
plt.bar(categories,values,bar_width,label='Dataset1')
plt.bar(categories,values2,bar_width=0.20,label='Dataset2')
plt.bar(categories,values3,bar_width=0.3,label='Dataset3')
plt.xlabel('categories')
plt.ylabel('values')
plt.title('Grouped bar chart Example')
plt.legend()
plt.show()
```

```python
explode=(0,0,0.1,0)
plt.pie(values,labels=categories,explode=explode)
plt.title('Exploded Pie Chart')
plt.show()
```

```python
explode=(0,0,0.1,0)
plt.pie(values,labels=categories,explode=explode,autopct="%1.1f%%")
plt.title('Exploded Pie Chart')
plt.show()
```

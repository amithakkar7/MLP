```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('iris.csv')
```

```python
data.head(5)
```

```python
x=data['sepal.length']
y=data['variety']
plt.plot(x,y)
plt.show()
```

```python
x=data['variety'].iloc[:70]
y=data['sepal.length'].iloc[:70]
plt.plot(x,y,'b^--')
plt.show()
```

```python
x=data['sepal.length']
y=data['sepal.width']
plt.plot(x,y,"bo--",linewidth=0.4)
plt.show()
```

```python
plt.xlabel('Numbers')
plt.ylabel('variety')
plt.title('Different Varieties')
plt.plot(data.variety)
plt.show()
```

```python
# Bar Chart
x=data['variety']
y=data['sepal.width']
plt.bar(x,y)
plt.show()
```

```python
# Histogram
x=data['variety']
plt.hist(x,bins=50)
plt.show()
```

```python
plt.hist(data['sepal.length'],bins=50)
plt.show()
```

```python
plt.hist(data['petal.length'],bins=50, histtype='bar',rwidth=0.5)
plt.show()
```

```python
x=np.random.normal(100,5,200)
print(x)
plt.hist(x)
plt.show()
```

```python
# Pie Chart
y=np.array([10,20,30,40])
mylabels=["ten","twenty","thirty","forty"]
myexplode=[0.2,0.2,0.2,0.2]
plt.pie(y,labels=mylabels,explode=myexplode,startangle=90,shadow=True)
plt.show()
```

```python
x=data['sepal.length'].iloc[:5]
myexplode=[0.5,0.1,0.1,0.1,0.1]
plt.pie(x,explode=myexplode,labels=x,shadow=True,autopct="%1.1f%%")
plt.title('Pie Chart of sepal length')
plt.legend(title='Legend',loc='upper right')
plt.show()
```

```python
#Scatter Plot
x=data['petal.length']
y=data['petal.width']
plt.scatter(x,y)
plt.show()
```

```python
x=data['variety']
y=data['sepal.length']
x1=data['variety']
y1=data['sepal.width']
plt.scatter(x,y,c='b')
plt.scatter(x1,y1,c='r')
plt.title('Scattered Plot')
plt.xlabel('Variety')
plt.ylabel('Petal Width')
plt.legend(data['variety'])
plt.show()
```
